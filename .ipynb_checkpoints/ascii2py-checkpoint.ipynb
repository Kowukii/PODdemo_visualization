```python
x_coords = np.arange(0, 0.095, 0.001)
y_coords = np.arange(0, 0.042, 0.001)
x_coords, y_coords
```

```python
target_x, target_y = np.meshgrid(
    x_coords,
    y_coords,
)
target_x.shape, target_y.shape
grid = np.meshgrid(
    x_coords,
    y_coords
)
xin_grid, yin_grid = grid
```

```python
xin_grid = np.concatenate(xin_grid)
yin_grid = np.concatenate(yin_grid)
```

```python
x_grid = df['    x-coordinate'].astype(float)
y_grid = df['    y-coordinate'].astype(float)
plt.scatter(x_grid, y_grid, s=1, c='red', label='data')
plt.show()
```

```python
vort = df['   vorticity-mag'].astype(float)
v_m = df['velocity-magnitude'].astype(float)
v_a = df['  velocity-angle'].astype(float)
# 计算速度的 x 和 y 分量
vx = v_m * np.cos(v_a)
vy = v_m * np.sin(v_a)
```

```python
import cartopy.crs as ccrs
def plot(lons, lats, values):
    fig = plt.figure(figsize=(10, 5))
    ax = plt.axes(
        projection=ccrs.PlateCarree()
    )
    p = ax.pcolormesh(
        lons, lats, values,
        shading='nearest',
        cmap="coolwarm", 
        vmin=-40, vmax=40
    )
    fig.colorbar(p, fraction=0.046, pad=0.04)
    ax.coastlines()
    plt.show()
```

```python
import numpy as np
from scipy.interpolate import griddata
# 生成坐标点对
points = np.column_stack((x_grid, y_grid))
targetpoints = np.column_stack((xin_grid, yin_grid))
vxgrid = griddata(points, vx * 10, (target_x, target_y), method='nearest')
vygrid = griddata(points, vy * 10, (target_x, target_y), method='nearest')
# vxgrid = vxgrid.reshape(95, 42)
vxgrid.shape
# plt.figure()

# ax1 = plt.subplot2grid((2,2), (0,0), projection='3d')
# ax1.plot_surface(xin_grid, yin_grid, vxgrid, color = "c")
# plt.show()
plot(x_coords, y_coords, vxgrid)
plot(x_coords, y_coords, vygrid)
```

```python
from scipy.interpolate import griddata
grid_z0 = griddata(points, vx * 10, (target_x, target_y), method='nearest')
grid_z1 = griddata(points, vx * 10, (target_x, target_y), method='linear')
grid_z2 = griddata(points, vx * 10, (target_x, target_y), method='cubic')

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
plt.figure()

ax1 = plt.subplot2grid((2,2), (0,0), projection='3d')
ax1.plot_surface(target_x, target_y, grid_z0, color = "c")
ax1.set_xlim3d(np.min(target_x), np.max(target_x))
ax1.set_ylim3d(np.min(target_y), np.max(target_y))
ax1.set_zlim3d(np.min(grid_z0), np.max(grid_z0))
ax1.set_title('nearest')

ax2 = plt.subplot2grid((2,2), (0,1), projection='3d')
ax2.plot_surface(target_x, target_y, grid_z1, color = "c")
ax2.set_xlim3d(np.min(target_x), np.max(target_x))
ax2.set_ylim3d(np.min(target_y), np.max(target_y))
ax2.set_zlim3d(np.nanmin(grid_z1), np.nanmax(grid_z1))
ax2.set_title('linear')

ax3 = plt.subplot2grid((2,2), (1,0), projection='3d')
ax3.plot_surface(target_x, target_y, grid_z2, color = "r")
ax3.set_xlim3d(np.min(target_x), np.max(target_x))
ax3.set_ylim3d(np.min(target_y), np.max(target_y))
ax3.set_zlim3d(np.nanmin(grid_z2), np.nanmax(grid_z2))
ax3.set_title('cubic')

ax4 = plt.subplot2grid((2,2), (1,1), projection='3d')
ax4.scatter(points[:,0], points[:,1], vx,  c= "b", s=1)
ax4.set_xlim3d(np.min(points[:,0]), np.max(points[:,0]))
ax4.set_ylim3d(np.min(points[:,1]), np.max(points[:,1]))
ax4.set_zlim3d(np.min(vx), np.max(vx))
ax4.set_title('org_points')

plt.tight_layout()
plt.show()

print( "*" * 20)
```

```python
def compute_vorticity(u, v):
    # 计算速度场的梯度
    du_dy, du_dx = np.gradient(u)
    dv_dy, dv_dx = np.gradient(v)

    # 计算涡度
    vorticity = dv_dx - du_dy

    return vorticity
```

```python
vortFromSpeed = compute_vorticity(vxgrid, vygrid)
vortFromSpeed.shape
```

```python
# 可视化涡度矩阵
plt.imshow(vortFromSpeed, cmap='jet')
plt.colorbar()
plt.title('Vorticity')
plt.show()
```
