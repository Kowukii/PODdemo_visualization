```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
def toReal(x):
    if np.iscomplexobj(x):
        return float(np.real(x))


def matrix2real(matrix):
    return np.vectorize(toReal)(matrix)
```

```python
x = np.arange(0, 5.01, 0.01)  # 501 空间
m = len(x)
t = np.arange(0, 5.05, 0.05)  # 101 时间
N = len(t)
Fs = 1 / (t[1] - t[0]) # 采样频率
X, T = np.meshgrid(x, t)
```

```python
U_tx = 1.2 * np.exp(-0.5 * T) * np.sin(2 * np.pi * (X + 2 * T)) + 0.8 * np.exp(0.3 * T) * np.sin(2 * np.pi * (3 * X + 4 * T)) + 1.1 + 0.1 * np.random.rand(N, m)
U_tx
```

```python
# 计算X和Y矩阵
U_xt = U_tx.T
X = U_xt[:, 0:N-1]
Y = U_xt[:, 1:N]
X, np.size(X, 0), np.size(X, 1)
```

```python
Y, np.size(Y, 0), np.size(Y, 1)
```

```python
U, S, VT = np.linalg.svd(X)
S, len(S)
```

```python
Sd = np.diag(S)
Sd
```

```python
r = np.sum(Sd > 1e-6) # 删除奇异值约等于0的模态，防止计算发散
r
```

```python
U = U[:, 0:r]
U
```

```python
S = Sd[0:r, 0:r]
S
```

```python
V = VT.T[:, 0:r]
V
```

```python
# 求解矩阵A
A = U.T @ (Y @ V @ np.linalg.inv(S))
A, np.size(A, 0), np.size(A, 1)
```

```python
# 求矩阵A的特征值和特征向量
D, Om = np.linalg.eig(A)
Dd = np.diag(D) # 特征值转为向量形式
Dd, np.size(Dd, 0), np.size(Dd, 1)
```

```python
Phi = Y @ V @ np.linalg.inv(S) @ Om
Phi.shape
```

```python
# 求解初始状态b
b, residuals, rank, singular_values = np.linalg.lstsq(Phi, X[:, 0], rcond=None)
# b = np.linalg.solve(Phi, X[:, 0])
b, len(b)
```

```python
# 对模态进行排序
Q = np.vander(D, N-1, increasing=True) # 计算范德蒙矩阵
Q, np.size(Q, 0), np.size(Q, 1)
```

```python
# Time_DMD = np.outer(b.reshape((1, len(b))), Q)
Time_DMD = b.reshape(len(b), 1) * Q
Time_DMD, Time_DMD.shape
```

```python
# 对信号初始振幅排序，速度快，输出能量近似用振幅平方代替

# 求出所有模态的信号，并计算能量
Energy = np.zeros(np.size(Phi, 1))
Energy, len(Energy), np.size(Phi, 0)
```

```python
# Uxt_DMD_k = np.real(np.outer(Phi[:, 0], Time_DMD[0, :]))
Uxt_DMD_k = np.real(Phi[:, 0].reshape((Phi.shape[0], 1)) * Time_DMD[0, :].reshape((1, Time_DMD.shape[0])))
Uxt_DMD_k, Uxt_DMD_k.shape
```

```python
for i in range(np.size(Phi, 1)):
    Uxt_DMD_k = np.real(Phi[:, i].reshape((Phi.shape[0], 1)) * Time_DMD[i, :].reshape((1, Time_DMD.shape[0])))
    E_k = np.sum(np.sum(Uxt_DMD_k ** 2))
    Energy[i] = np.log(E_k)

Energy, len(Energy)
```

```python
# 对于每个模态的能量进行排序
Ie = np.argsort(-Energy)
Energy = Energy[Ie]
Energy, len(Energy)
```

```python
Dd = Dd[Ie]
Dd
```

```python
b = b[Ie]
b
```

```python
Phi = Phi[:, Ie]
Phi.shape, Ie.shape
```

```python
Time_DMD = Time_DMD[Ie, :]
Time_DMD.shape
```

```python
def DMD_class(X, Y):
    #DMD分解函数
    # 输入： X， Y， DMD分解的数据矩阵
    # 输出： 
    #      Dd, 特征根
    #       b, 初始状态
    #     Phi, DMD分解的模态
    #Time_DMD, DMD还原信号所用到的时间项
    #  Energy, 每个模态对应的能量，从大到小排序
    N = np.size(X, 1)
    # SVD分解
    U, S, VT = np.linalg.svd(X, full_matrices=False)
    Sd = np.diag(S)
    r = np.sum(Sd > 1e-6) # 删除奇异值约等于0的模态，防止计算发散
    U = U[:, 0:r]
    S = Sd[0:r, 0:r]
    V = VT.T[:, 0:r]
    # 求解矩阵A
    A = U.T @ (Y @ V @ np.linalg.inv(S))
    # 求矩阵A的特征值和特征向量
    D, Om = np.linalg.eig(A)
    Dd = np.diag(D) # 特征值转为向量形式
    # 求DMD模态
    Phi = Y @ V @ scipy.linalg.inv(S) @ Om
    # 求解初始状态b
    b, residuals, rank, singular_values = np.linalg.lstsq(Phi, X[:, 0], rcond=None)
    # b = np.linalg.solve(Phi, X[:, 0])
    # 对模态进行排序
    Q = np.vander(D, N, increasing=True) # 计算范德蒙矩阵
    # Time_DMD = np.outer(b, Q)
    Time_DMD = b.reshape(len(b), 1) * Q
    # 对信号初始振幅排序，速度快，输出能量近似用振幅平方代替
    
    # 求出所有模态的信号，并计算能量
    Energy = np.zeros(np.size(Phi, 1))
    for i in range(np.size(Phi, 1)):
        Uxt_DMD_k = np.real(Phi[:, i].reshape((Phi.shape[0], 1)) * Time_DMD[i, :].reshape((1, Time_DMD.shape[0])))
        E_k = np.sum((Uxt_DMD_k ** 2))
        Energy[i] = E_k

    # 对于每个模态的能量进行排序
    Ie = np.argsort(-Energy)

    Energy = Energy[Ie]
    Dd = Dd[Ie]
    b = b[Ie]
    Phi = Phi[:, Ie]
    Time_DMD = Time_DMD[Ie, :]

    
    return Dd, b, Phi, Time_DMD, Energy
    
```

```python
# 计算DMD
Dd, b, Phi, Time_DMD, Energy = DMD_class(X, Y) 
```

```python
fig = plt.figure(1)
# 图1 输出特征根分布
plt.scatter(np.sum(np.real(Dd[::-1]), axis=1),
            np.sum(np.imag(Dd[::-1]), axis=1),
            s=30,
            c=-np.log(Energy[::-1]),
            cmap='viridis',
            alpha=0.7)
plt.colorbar()
plt.axis('equal')
plt.show()
```

```python
fig2 = plt.figure(2)
# 图2 绘制频率和衰减图
wa = np.log(np.sum(Dd, axis=1)) * Fs
# 增加一个极小的数防止计算过程出现inf
wa, len(wa)
```

```python
plt.scatter(np.real(wa),
            np.imag(wa) / (2 * np.pi),
            s=30,
            c=-np.log(Energy),
            cmap='viridis',
            alpha=0.7)
plt.xlabel('AttenuationRatio σ')
plt.ylabel('Frequency w')
plt.colorbar()
plt.xlim(-1, 1)
plt.ylim(-6, 6)
plt.show()
```

```python
fig3 = plt.figure(3)
# 频率-能量排序
Freq = np.imag(wa) / 2 / np.pi
Freq
```

```python
# 找到大于等于0的频率部分的索引
k = np.where(Freq >= 0)[0]

# 绘制幅度谱
plt.stem(Freq[k], np.log10(Energy[k]), basefmt='b', linefmt='b-', 
         markerfmt='bo')

# 设置基线值为-6
plt.axhline(y=-4, color='r', linestyle='--')

# 显示图形
plt.show()
```
