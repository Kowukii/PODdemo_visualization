```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def toReal(x):
    if np.iscomplexobj(x):
        return float(np.real(x))


def matrix2real(matrix):
    return np.vectorize(toReal)(matrix)
```

```python
x = np.arange(0, 5.01, 0.01)  # 501
t = np.arange(0, 5.01, 0.1)  # 51
N = len(t)
Fs = 1 / (t[1] - t[0])
X, T = np.meshgrid(x, t)
```

```python
Utx = 1.2 * np.exp(- 0.5 * T) * np.sin(2 * np.pi * (X + 0 * T)) + 0.8 * np.exp(0.3 * T) * np.sin(2 * np.pi * (3 * X + 4 * T))
np.size(Utx, 0), np.size(Utx, 1)
# 原始信号
```

```python
Utx
```

```python
Uxt = Utx.T 
np.size(Uxt, 0), np.size(Uxt, 1)
# 时间 * 空间的矩阵
```

```python
Uxt
```

```python
Up = np.linalg.pinv(Uxt[:, 0:50])
np.size(Up, 0), np.size(Up, 1)
# 求解X的伪逆矩阵
```

```python
Up
```

```python
A = Uxt[:, 1:51] @ Up
np.size(A, 0), np.size(A, 1)
```

```python
A
```

```python
D, V = np.linalg.eig(A)
D
```

```python
# Ds = np.diag(D)
ind = np.argsort(-np.abs(D)) # 对D降序排列
```

```python
ind
```

```python
# Ds = Ds[ind]
D = D[ind]
V = V[:, ind]
```

```python
# Dabs = np.abs(Ds) + np.ones((np.size(Ds, 0), np.size(Ds, 1)))
Dabs = np.abs(D)
Dd = np.divide(D, Dabs)
```

```python
Dabs
```

```python
Dd
```

```python
# 频率
f = Fs * np.angle(Dd) / 2 * np.pi
f
```

```python
rValue = np.real(Dd[99::-1])
rValue
```

```python
iValue = np.imag(Dd[99::-1])
iValue
```

```python
colorMap = -np.abs(D[99::-1])
# 根据绝对值的负值上色
len(colorMap)         
```

```python
fig = plt.figure(1)
plt.scatter(rValue,
            iValue,
            s=10,
            c=colorMap,
            alpha=0.7)
plt.colorbar()
plt.axis('equal')
plt.show()
```
